## One Way ANOVA

```python
df.columns
```

```python
import pandas as pd
from scipy import stats

# Filter relevant columns
techniques = df['Which project management methodologies are primarily used in your company? (Check all that apply)']
cycle_times = df['What is the current average production cycle time after the implementation of project management techniques? (in )']

# Group by technique and prepare for ANOVA
groups = [cycle_times[techniques == technique] for technique in techniques.unique()]

# Perform ANOVA
f_statistic, p_value = stats.f_oneway(*groups)
print(f'F-statistic: {f_statistic}, p-value: {p_value}')
```
